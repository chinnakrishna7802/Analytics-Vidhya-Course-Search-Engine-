```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# Scraping the data

```python
corpus=BeautifulSoup()
n=8

def scrape_courses(url):
    for i in range(1,n+1):
        response = requests.get(url+'?page='+str(i))
        soup = BeautifulSoup(response.content, 'html.parser')
        corpus.append(soup)
    courses = corpus.find_all('a', class_='course-card course-card__public published')
    course_dict={}
    corpus_d=BeautifulSoup()
    for course in courses:
        title = course.find('h3').text.strip()
        href = course.get('href')
        response = requests.get('https://courses.analyticsvidhya.com'+href)
        soup = BeautifulSoup(response.content, 'html.parser')
        course_fee = soup.find('span', class_='course-card__price')

        if course_fee:
            continue
        else:
            if soup.find('div', class_='custom-theme'):
                course_description = soup.find('div', class_='custom-theme').find(['p','span']).text.strip()
                curriculum = []
                chapters = soup.find_all('li', class_='course-curriculum__chapter')

                for chapter in chapters:
                    chapter_title = chapter.find('h5', class_='course-curriculum__chapter-title').text.strip()
                    lessons = [lesson.text.strip() for lesson in chapter.find_all('span', class_='course-curriculum__chapter-lesson')]
                    curriculum.append({'chapter': chapter_title, 'lessons': lessons})

                
                course_dict[title]={'Tile':title,'course_description':course_description,'curriculum':curriculum,'href':'https://courses.analyticsvidhya.com'+href}

    return course_dict
course_dict=scrape_courses('https://courses.analyticsvidhya.com/collections')
        
```

# Convert to Dataframe

```python
import pandas as pd
data = pd.DataFrame(course_dict).transpose()
data
```

```python

def curr(l):
    str=''
    for item in l:
        str += f"Chapter: {item['chapter']}\nLessons: {', '.join(item['lessons'])}\n\n"
    return str

data['curriculum']=data['curriculum'].apply(curr)

```

```python
data['text']=data['Tile']+data['course_description']
data.to_csv('data.csv')
```

# Embedding data using langchain and huggingface


```python
import os
from dotenv import load_dotenv
load_dotenv()  #load all the environment variables
```

```python
os.environ['HF_TOKEN']=os.getenv("HF_TOKEN")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
type(data['embedding'].iloc[0])
```

```python

# embedded_data = embeddings.embed(data['text'].tolist())

# Convert the embedded data to a Pandas DataFrame
data['embedding'] = data['text'].apply(lambda text: embeddings.embed_query(text))

data[['Tile', 'course_description', 'embedding']]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
def recommend_courses(query, df , top=3):
    # Generate embedding for the input query
    query_embedding = embeddings.embed_query(query)

    # Calculate cosine similarity between the query embedding and course embeddings
    df['similarity'] = df['embedding'].apply(lambda x: cosine_similarity([query_embedding], [x])[0][0])

    # Sort the DataFrame by similarity score in descending order and get the top_n results
    recommended_courses = df.sort_values(by='similarity', ascending=False).head(top)
    
    return recommended_courses[['Tile', 'course_description', 'similarity','href']]

recommend_courses('machine learning' , data , top=3)
```
